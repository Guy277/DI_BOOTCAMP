=== INTRODUCTION TO DATA ANALYSIS ===

1. WHAT IS DATA ANALYSIS?
Data analysis is the process of collecting, cleaning, and interpreting data
to extract useful information and support decision-making. It combines
statistics, logic, and tools to turn raw numbers into meaningful insights.

2. WHY IS IT IMPORTANT?
In today's world, data is generated every second — from phones, hospitals,
banks, and cities. Without analysis, this data is just noise. Data analysis
helps businesses, governments, and researchers make smarter, evidence-based
decisions instead of guessing.

3. THREE AREAS WHERE DATA ANALYSIS IS APPLIED:

   a) Healthcare
      Doctors and researchers analyse patient data to detect diseases early,
      track epidemics, and personalise treatments. Example: COVID-19 dashboards
      used real-time data to guide public health decisions worldwide.

   b) Finance
      Banks use data analysis to detect fraud, evaluate credit risk, and guide
      investments. Algorithms analyse thousands of transactions per second to
      flag suspicious activity instantly.

   c) Urban Planning
      Cities analyse traffic, energy consumption, and population data to build
      smarter infrastructure. In Africa, satellite and mobile data help target
      development aid and improve agricultural planning.

A column is **quantitative** when it contains a measurable numeric value.  
A column is **qualitative** when it represents a category, label, code, or class.

## 1. Time Americans Spend Sleeping

- `Year` : **Quantitative**. It is numeric and the difference between years is meaningful.
- `Period` : **Qualitative**. It is a label such as `Annual`.
- `Avg hrs per day sleeping` : **Quantitative**. It measures hours of sleep.
- `Standard Error` : **Quantitative**. It is a numerical statistical measure.
- `Type of Days` : **Qualitative**. It describes categories of days.
- `Age Group` : **Qualitative**. It is a grouped category, not an exact measurement.

# Compétences de base en observation et analyse de données

## Colonnes pertinentes pour une analyse de tendance

- `Year` : cette colonne est importante pour observer l’évolution du sommeil au fil du temps.
- `Avg hrs per day sleeping` : cette colonne est la plus importante, car elle montre la moyenne d’heures de sommeil par jour.
- `Standard Error` : cette colonne est utile pour savoir si les résultats sont plus ou moins précis.

Exemple de question :
Comment le nombre moyen d’heures de sommeil a-t-il évolué entre 2003 et 2017 ?

## Colonnes pertinentes pour une comparaison de groupes

- `Age Group` : utile pour comparer le sommeil selon les groupes d’âge.
- `Sex` : utile pour comparer le sommeil entre hommes, femmes, ou les deux.
- `Type of Days` : utile pour comparer le sommeil selon le type de jour, par exemple jours de semaine ou week-end.
- `Avg hrs per day sleeping` : nécessaire pour mesurer la quantité de sommeil dans chaque groupe.

Exemple de question :
Les jeunes dorment-ils plus que les personnes âgées ?
Les Américains dorment-ils plus le week-end que pendant les jours de semaine ?

## Colonnes moins utiles pour cette analyse

- `index` : ce n’est qu’un identifiant de ligne, donc cette colonne n’apporte pas d’information analytique.
- `Period` : cette colonne n’est pas très utile ici car elle contient seulement `Annual`.
- `Activity` : cette colonne n’est pas très utile ici car elle contient seulement `Sleeping`.
